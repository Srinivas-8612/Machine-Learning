```python
import torch
import torch.nn as nn

class PipelineModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.stage1 = nn.Linear(4, 8)
        self.stage2 = nn.ReLU()
        self.stage3 = nn.Linear(8, 2)

    def forward(self, x):
        print("Input:", x)
        print("Input shape:", tuple(x.shape))

        x = self.stage1(x)
        print("\nAfter Stage 1 (Linear):")
        print("Activation:", x)
        print("Shape:", tuple(x.shape))

        x = self.stage2(x)
        print("\nAfter Stage 2 (ReLU):")
        print("Activation:", x)
        print("Shape:", tuple(x.shape))

        x = self.stage3(x)
        print("\nAfter Stage 3 (Linear):")
        print("Activation:", x)
        print("Shape:", tuple(x.shape))

        return x


model = PipelineModel()
x = torch.randn(1, 4)

print("Running model...")
output = model(x)

print("\nFinal Output:", output)
print("Final Shape:", tuple(output.shape))
```
